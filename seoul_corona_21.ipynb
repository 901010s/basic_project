# 데이터 불러오기

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('서울시 코로나19 확진자 현황.csv', low_memory=False)
df
```

# 데이터 확인 및 전처리

컬럼별 데이터 확인

```python
df['조치사항'].unique()
```

```python
df.columns
```

```python
df.isnull().sum()
```

불필요한 컬럼 삭제

```python
df.drop(columns=['환자번호', '국적', '환자정보',
                 '조치사항', '이동경로', '등록일',
                 '수정일', '노출여부'], inplace=True)
df
```

```python
df.dtypes
```

```python
# 확진일 -> datetime
df['확진일'] = pd.to_datetime(df['확진일'])
```

```python
df.dtypes
```

```python
# 지역의 공백 제거
df['지역'].nunique()
```

```python
df['지역'].unique()
```

```python
df['지역'] = df['지역'].str.strip()
```

```python
df['지역'].nunique()
```

```python
df['지역'].unique()
```

```python
# 지역, 타입을 category로 변경
df['지역'] = df['지역'].astype('category')
```

```python
df.info()
```

# 구별 확진자 동향
확진일 - 구별 확진자수 집계 pivotable 만들기

```python
df_gu = pd.pivot_table(df, index='확진일', columns='지역',
                       values='연번', aggfunc='count',
                       margins=True)
df_gu
```

# 서울시 일별 추가 확진자 동향

```python
s_date = df_gu['All'][:-1]
s_date
```

```python
# 일별 추가 확진자가 많았던 순
s_date.sort_values(ascending=False)
```

```python
# 서울시 일별 추가 확진자 시각화
```

```python
x = s_date.index
y = s_date.values
plt.plot(x, y)
plt.title('서울시 일별 추가 확진자 시각화')
plt.xlabel('확진일')
plt.ylabel('확진자수')
plt.xticks(rotation=45)
plt.show()
```

# 서울 구별 누적확진자 비교

```python
# 서울시 구별 누적 확진자가 많은 순으로
s_gu = df_gu.loc['All'][:-1]
s_gu
```

```python
s_gu = s_gu.sort_values(ascending=False)
s_gu
```

```python
# 서울시 구별 누적 확진자 많은순 시각화
x = s_gu.index
y = s_gu.values
plt.figure(figsize=(10,8))
plt.barh(x, y)
plt.title('서울시 구별 누적확진자수', size = 10)
plt.xlabel('확진자수')
plt.ylabel('자치구')
plt.show()
```
